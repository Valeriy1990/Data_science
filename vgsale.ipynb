# Задание 1

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('/content/vgsale_1.csv')
df
```

```python
df_n = df[(df['Year'].notnull())]

print()
print(f'Общее количество игр за период (Year != NaN): {df[(df["Year"].notnull())]["Name"].nunique()}')
print()

df_n = df_n.drop_duplicates(subset=['Name'])
df_n.info()

ax = sns.histplot(df_n["Genre"], discrete=True)

ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
df_left = df_n[(df_n.Year <= 2000.0)]

df_left.info()
print()
print(f'Игры выпущенные до 2000 года (включительно) (Year != NaN): {df_left["Name"].nunique()}')
print()
df_left

ax = sns.histplot(df_left['Genre'], discrete=True)

ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

 До 2000 года, самыми популярными были игры спортивного жанра.

```python
df_right = df_n[(df_n.Year >= 2000.0)]

df_right.info()
print()
print(f'Игры выпущенные после 2000 года (включительно) (Year != NaN): {df_right["Name"].nunique()}')
print()

ax = sns.histplot(df_right['Genre'], discrete=True)

ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

После 2000 года, самыми популярными были игры жанра 'Action'.

```python
df_Global = df[(df['Name'].notnull())]

df_Global = df_Global[['Genre', 'Global_Sales']].groupby('Genre').aggregate('sum')

df_Global.sort_values('Global_Sales', ascending=False).plot(kind='bar')

plt.show()
```

По показателю "Объёмы продаж", за весь период, лучше всего себя показали игры жанра "Action"

```python
df_Yers= df[(df['Year'].notnull())]

df_Yers['Year'] = df_Yers['Year'].astype('int')

df_Yers = df_Yers.drop_duplicates(subset=['Name'])

df_Yers.info()

plt.figure(figsize= (12, 4))
df_Yers_groupby = df_Yers.groupby('Year').aggregate('count')

df_Yers_groupby['Name'].plot(kind='bar', rot = 45)
plt.tight_layout()
plt.show()

m = df_Yers_groupby[(df_Yers_groupby['Name'] == df_Yers_groupby['Name'].max())]
print(f'Больше всего выпущено игр в {int(m.index[0])} году, в колличестве {int(m.iloc[0,0])}')
```

```python
df_Publisher= df[(df['Publisher'].notnull())]
df_Publisher = df_Publisher.drop_duplicates(subset=['Name'])

df_Publisher.info()

df_Publisher = df_Publisher.groupby('Publisher').aggregate('count').sort_values('Name', ascending=False)
df_Publisher['Name'].head(3).plot(kind='bar', rot = 45)
plt.tight_layout()
plt.show()

m = df_Publisher[(df_Publisher['Name'] == df_Publisher['Name'].max())]
print(f'Больше всего игр, за указанный период, выпущено компанией {m.index[0]}, в колличестве {int(m.iloc[0,0])}')
```

```python
df_Sales= df[(df['Year'].notnull())]

df_Sales['Year'] = df_Sales['Year'].astype('int')

df_Sales.info()

my_colum = ['Other_Sales', 'JP_Sales', 'EU_Sales', 'NA_Sales']

df_Sales_1980_2000 = df_Sales.query("Year >= 1980 & Year <= 2000")[my_colum].sum()

df_Sales_2000_2020 = df_Sales.query("Year >= 2000 & Year <= 2020")[my_colum].sum()

fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.pie(df_Sales_1980_2000.values, labels=my_colum, autopct='%1.1f%%')
ax1.set_title("1980 - 2000 гг.")
ax2.pie(df_Sales_2000_2020.values, labels=my_colum, autopct='%1.1f%%')
ax2.set_title("2000 - 2020 гг.")
plt.show()

print()
print("Доли суммарного объема продаж с 1980 г. до 2000 г. и с 2000 г. до 2020")
```

# Задание 2

```python
df = pd.read_csv('/content/IQ_countries.csv')
df
```

```python
df.info()
df['Population'] = df['Population'].astype('float64')
df_mean = df.mean(numeric_only=True)
df_mean
```

```python
# Меняю все NaN на средние значения для pairplot

df_float = df.fillna(df_mean)

df_float.info()
```

```python
df_float = df[['Average IQ', 'Literacy Rate', 'Human Development Index', 'Mean years of schooling', 'Gross National Income', 'Population']]
sns.pairplot(df_float, hue='Average IQ')
```

```python
# Оцениваем взаимосвязь факторов между собой

f,ax = plt.subplots(1,1)
sns.heatmap(df_float.corr(), annot=True, square=True, cmap='coolwarm')
plt.show()
plt.close()
```

```python
df_n = df.drop(['Rank', 'Country', 'Population'], axis=1).groupby("Continent").aggregate("mean").sort_values('Average IQ', ascending=False)
df_n
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize = (16, 10))
ax1.plot('Literacy Rate', 'Average IQ', data=df_n)
ax1.set_ylabel('Average IQ')
ax1.set_xlabel('Literacy Rate')

ax2.plot('Human Development Index', 'Average IQ', data=df_n)
ax2.set_xlabel('Human Development Index')

ax3.plot('Mean years of schooling', 'Average IQ', data=df_n)
ax3.set_xlabel('Mean years of schooling')

ax4.plot('Gross National Income', 'Average IQ', data=df_n)
ax4.set_xlabel('Gross National Income')

fig.autofmt_xdate(rotation=65)
plt.show()
```

```python
df_n = df.drop(['Rank', 'Continent', 'Population'], axis=1).groupby("Country").aggregate("mean").sort_values('Average IQ', ascending=False)
df_n
```

В целом можно сделать следующие выводы:
- Средний показатель IQ положительно коррелирует с коэффициентом грамотности (Чем больше грамотность, тем выше IQ)
- Средний показатель IQ положительно коррелирует с показателем “валовой национальный доход” и среднее количество лет школы(Чем больше “валовой национальный доход” и среднее количество лет школы, тем выше IQ)
- Индекс человеческого развития положительно коррелирует со средним показателем IQ (Чем больше средний показатель IQ, тем выше индекс человеческого развития)

- Положительно коррелируют между собой коэффициент грамотности и среднее количество лет школы.
- Положительно коррелируют между собой показатели индекса человеческого развития и коэффициент грамотности.

- Валовой национальный доход положительно коррелирует с индексом человеческого развития.

- Самый большой показатель IQ на европейском континенте.
- Самый большой показатель IQ в Японии.
- Не один из показателей не коррелирует от численности населения.


# Задание 3

```python
df = pd.read_csv('/content/shopping_habits.csv')
df
```

```python
df.info()

df.groupby('Gender').aggregate('describe').T
```

```python
ax = sns.histplot(df['Category'], discrete=True)

ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
column = ['Accessories', 'Clothing', 'Footwear', 'Outerwear']

df_Fall = df.query("Season == 'Fall'")[['Category', 'Customer ID']].groupby(['Category'], as_index = False).aggregate('count')
df_Spring = df.query("Season == 'Spring'")[['Category', 'Customer ID']].groupby(['Category'], as_index = False).aggregate('count')
df_Summer = df.query("Season == 'Summer'")[['Category', 'Customer ID']].groupby(['Category'], as_index = False).aggregate('count')
df_Winter = df.query("Season == 'Winter'")[['Category', 'Customer ID']].groupby(['Category'], as_index = False).aggregate('count')

plt.figure(figsize= (20, 20))

fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize = (20, 16))
ax1.pie(df_Fall['Customer ID'], labels=column, autopct='%1.1f%%')
ax1.set_title("Fall")

ax2.pie(df_Spring['Customer ID'], labels=column, autopct='%1.1f%%')
ax2.set_title("Spring")

ax3.pie(df_Summer['Customer ID'], labels=column, autopct='%1.1f%%')
ax3.set_title("Summer")

ax4.pie(df_Winter['Customer ID'], labels=column, autopct='%1.1f%%')
ax4.set_title("Winter")



plt.show()
```

```python
df[['Gender', 'Size', 'Customer ID']].groupby(['Gender', 'Size']).aggregate('count')
```

```python
df_Payment_Method = df[['Payment Method', 'Customer ID']].groupby(['Payment Method'], as_index = False).aggregate('count')

column = ['Bank Transfer', 'Cash', 'Credit Card', 'Debit Card', 'PayPal', 'Venmo']
plt.pie(df_Payment_Method['Customer ID'], labels=column, autopct='%1.1f%%')
plt.title("Payment_Method")

plt.show()
```
